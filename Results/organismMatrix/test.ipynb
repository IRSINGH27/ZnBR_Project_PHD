```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from numpy import log10
```

```python
df=pd.read_csv('organismMatrix.tsv',sep='\t',index_col=0)
```

```python
df['logznbr']=df.ZnBR_Count.apply(log10)
df['logpLen']=df.pLen.apply(log10)
df['znbr/pLen']=df.ZnBR_Count/df.pLen
df
```

```python
sbn.relplot(df,y='ZnBR_Count',x='pLen',hue='superkingdom',col='superkingdom')
sbn.catplot(df,y='ZnBR_Count',x='superkingdom',hue='superkingdom',kind='box')
sbn.catplot(df,y='znbr/pLen',hue='superkingdom',x='superkingdom',kind='box')

sbn.relplot(df,y='logznbr',x='logpLen',hue='superkingdom',col='superkingdom')
```

```python
ogDF=pd.read_csv('binaryCount/eggNOG_mmseq2_ZnBROG_binary.tsv',sep='\t',index_col=0)
```

```python
df.loc[:,'eggNOG_b']=ogDF.sum(axis=1).loc[df.index].values
```

```python
from scipy.stats import pearsonr,linregress
```

```python
x=df['eggNOG'].values
y=df['ZnBR_Count'].values
print(pearsonr(x,y))
print(linregress(x,y))
```

```python
x=df.loc[df.superkingdom=='Archaea','eggNOG'].values
y=df.loc[df.superkingdom=='Archaea','ZnBR_Count'].values
print(pearsonr(x,y))
print(linregress(x,y))
```

```python
x=df.loc[df.superkingdom!='Archaea','eggNOG'].values
y=df.loc[df.superkingdom!='Archaea','ZnBR_Count'].values
print(pearsonr(x,y))
print(linregress(x,y))
```
