```r
df2<-read.csv('16s_fullinfo.tsv',sep='\t',row.names=1)
df2<-df2[,c('superkingdom','phylum','class','order','family','genus','species')]
x<-unique(df2$phylum)
for (i in x){
    df<-df2
    df[df$phylum==x[18],'phylum']<-'-'
    df[df$phylum!=i,'phylum']<-NA
    df$node<-treeio::nodeid(tree,row.names(df))
    tree2<-treeio::full_join(tree,df,by='node')
    p<-ggtree(tree2,color='grey')+geom_tippoint(aes(color=phylum))+scale_color_discrete(na.value='transparent')
    print(p)

}
```
